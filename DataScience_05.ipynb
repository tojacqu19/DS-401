```python
def count_different_tokens(ftokens):
    list_diff = []
    for i in ftokens:
        if i not in list_diff:
            list_diff.append(i)
    return list_diff
count_different_tokens(ftokens)
```

```python
print(len(count_different_tokens(ftokens)))
```

```python
import matplotlib.pyplot as plt

count = 0
occurence = []
def plot_tokens(count_different_tokens, ftokens):
    for i in count_different_tokens(ftokens): 
        for j in ftokens:
            if i==j:
                count = count+ 1
        occurence.append(count)
        count = 0
    plot_list = plt.bar(ftokens, occurence )
    return plot_list
plot_tokens(count_different_tokens, ftokens)
```
